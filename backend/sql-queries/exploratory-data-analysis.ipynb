## Emails Sent and Received (to Self and Others)

```sql
-- All Emails Self

-- Emails Sent to Others

    SELECT SenderID, RecepientID, COUNT(*) as EmailsSent 
        FROM tbl_ex_EmailAddresses
        GROUP BY SenderID, RecepientID
        ORDER BY EmailsSent DESC
```

```sql
-- Emails Sent to Self

    SELECT SenderID, RecepientID, COUNT(*) as EmailsSent 
        FROM tbl_ex_EmailAddresses
        GROUP BY SenderID, RecepientID
        HAVING SenderID = RecepientID
        ORDER BY EmailsSent DESC
```

```sql
-- Emails Sent to Others

    SELECT SenderID, RecepientID, COUNT(*) as EmailsSent 
        FROM tbl_ex_EmailAddresses
        GROUP BY SenderID, RecepientID
        HAVING SenderID != RecepientID
        ORDER BY EmailsSent DESC
```

## All Users

```sql
-- The idea of this query is that if the dataset is very big,
-- and the users table has a lot of  people who are not involved in the communications, 
-- this query is supposed to select all of the people to make a subset of the users table
-- such that it only contains the users who have participated in the conversation 
-- in a given dataset

WITH InvolvedPeople AS (
      SELECT SenderID, RecepientID as EmailsSent 
        FROM tbl_ex_EmailAddresses
        GROUP BY SenderID, RecepientID
        -- ORDER BY EmailsSent DESC
    ) 
    
    SELECT SenderID FROM InvolvedPeople

    UNION

    SELECT RecepientID FROM InvolvedPeople
```
